```python
import pprint as pp
import re
```

```python
bb = {}
name = re.compile(r'([a-zA-Z&\(\)\s,]+)\s+[0-9,\s]+')
with open('BPG Index.txt','r') as bf:
    for line in bf:
        line = line.strip()
        print(line)
        index = line[0]
        m = name.match(line)
        if m:
            if index in bb:
                bb[index].append(m.group(1))
            else:
                bb[index] = [m.group(1)]
  
```

```python
  pp.pprint(bb['B'])
```

```python
names = [bb[key][i] for key in bb.keys() for i in range(len(bb[key]))]
indexes = [name[0] for name in names]
```

```python
import pandas as pd
import numpy as np
from pandasql import sqldf
```

```python
d = {"Name": names, "City": 'Sacramento','Index_value':indexes}
df = pd.DataFrame(d)
df.head(10)
```

```python
def pyslqdf(q):
    return sqldf(q,globals())
```

# What are the number of entries in the Dataset by Alphabet Index?

```python
q = """
SELECT Index_value,
Count(Index_value) AS "Count"
   
FROM
    df a
GROUP BY 
    Index_value;
"""

count_df = pyslqdf(q)
count_df
```
